```python
4. Write a program that checks if a given word is a Python keyword.¬†

Ans. import keyword

def is_python_keyword(word):
  """
  Checks if a given word is a Python keyword.

  Args:
    word: The string to check.

  Returns:
    True if the word is a Python keyword, False otherwise.
  """
  return word in keyword.kwlist

# --- Main part of the program ---
if __name__ == "__main__":
  user_word = input("Enter a word to check if it's a Python keyword: ")

  if is_python_keyword(user_word):
    print(f"'{user_word}' is a Python keyword. üêç")
  else:
    print(f"'{user_word}' is NOT a Python keyword.")

  print("\n--- Testing with some examples ---")
  print(f"'if' is a keyword: {is_python_keyword('if')}")
  print(f"'else' is a keyword: {is_python_keyword('else')}")
  print(f"'my_variable' is a keyword: {is_python_keyword('my_variable')}")
  print(f"'True' is a keyword: {is_python_keyword('True')}")
  print(f"'true' is a keyword: {is_python_keyword('true')}") # Case-sensitive    
```

```python
5. Create a list and tuple in Python, and demonstrate how attempting to change an element works differently¬†for each

Ans.# Create a list
my_list = [10, 20, 30, 40, 50]
print(f"Original List: {my_list}")

# Attempting to change an element in the list
my_list[2] = 35 # Change the element at index 2 (which was 30)
print(f"List after changing element at index 2: {my_list}")

# You can also add elements
my_list.append(60)
print(f"List after appending 60: {my_list}")

# You can also remove elements
my_list.remove(10)
print(f"List after removing 10: {my_list}")

Explanation: As seen in the output, the element at index 2 in my_list was successfully changed from 30 to 35. Elements were also appended and removed without issue.

# Create a tuple
my_tuple = (100, 200, 300, 400, 500)
print(f"Original Tuple: {my_tuple}")

# Attempting to change an element in the tuple
try:
    my_tuple[2] = 350 # This line will cause an error
except TypeError as e:
    print(f"Error attempting to change element in tuple: {e}")

# Attempting to add an element
try:
    my_tuple.append(600) # This line will cause an error
except AttributeError as e:
    print(f"Error attempting to append to tuple: {e}")

# Attempting to remove an element
try:
    my_tuple.remove(100) # This line will cause an error
except AttributeError as e:
    print(f"Error attempting to remove from tuple: {e}")

print(f"Tuple remains unchanged: {my_tuple}") # The tuple's content will be the same as original

Explanation: When you try to reassign an element my_tuple[2] = 350, Python raises a TypeError because tuple objects "do not support item assignment."
```

```python
6. Write a function to demonstrate the behavior of mutable and immutable arguments.
    
Ans. demonstrate_argument_behavior(immutable_arg, mutable_arg):
    """
    Demonstrates how mutable and immutable arguments behave inside a function.

    Args:
        immutable_arg: An immutable object (e.g., int, str, tuple).
        mutable_arg: A mutable object (e.g., list, dict, set).
    """
    print(f"--- Inside the function ---")

    # --- Immutable Argument ---
    print(f"Immutable argument (initial): {immutable_arg}, ID: {id(immutable_arg)}")
    immutable_arg = immutable_arg + " changed" # This creates a NEW string object
    print(f"Immutable argument (modified): {immutable_arg}, ID: {id(immutable_arg)}")
    print("Notice how the ID changed for the immutable argument inside the function.")

    print("\n")

    # --- Mutable Argument ---
    print(f"Mutable argument (initial): {mutable_arg}, ID: {id(mutable_arg)}")
    mutable_arg.append(4) # This modifies the ORIGINAL list object
    mutable_arg[0] = 99   # This modifies the ORIGINAL list object
    print(f"Mutable argument (modified): {mutable_arg}, ID: {id(mutable_arg)}")
    print("Notice how the ID remained the same for the mutable argument inside the function.")
    print("This means the original object was modified directly.")

# --- Outside the function (Global Scope) ---
print("--- Outside the function (Global Scope) ---")

my_number = 10
my_string = "hello"
my_list = [1, 2, 3]

print(f"Original number: {my_number}, ID: {id(my_number)}")
print(f"Original string: '{my_string}', ID: {id(my_string)}")
print(f"Original list: {my_list}, ID: {id(my_list)}")

print("\nCalling the function with these arguments...")
demonstrate_argument_behavior(my_string, my_list)

print("\n--- After the function call (Global Scope) ---")

print(f"Number after function call: {my_number}, ID: {id(my_number)}")
print(f"String after function call: '{my_string}', ID: {id(my_string)}")
print(f"List after function call: {my_list}, ID: {id(my_list)}")

Explanation of Output
When you run the code, you'll observe:

For my_string (Immutable):

Inside the function, when immutable_arg is reassigned (immutable_arg = immutable_arg + " changed"), its ID changes. This indicates that a completely new string object was created within the function's scope.

Outside the function, my_string retains its original value ("hello") and its original ID. The function's internal change did not affect the global my_string.

For my_list (Mutable):

Inside the function, when mutable_arg is modified using methods like append() or item assignment (mutable_arg[0] = 99), its ID remains the same. This confirms that the function is operating directly on the original list object passed from the global scope.

Outside the function, my_list reflects these changes ([99, 2, 3, 4]). The modifications made inside the function are visible in the global scope.

This demonstration clearly shows that changes to mutable arguments persist outside the function, while changes (or rather, reassignments) to immutable arguments do not.
```

```python
7. Write a program that performs basic arithmetic operations on two user-input numbers.
    
Ans. perform_calculations():
    """
    Prompts the user for two numbers and performs addition, subtraction,
    multiplication, and division.
    """
    print("Welcome to the Basic Arithmetic Calculator!")

    # Get the first number from the user
    while True:
        try:
            num1_str = input("Enter the first number: ")
            num1 = float(num1_str) # Convert input to a float to handle decimals
            break # Exit loop if conversion is successful
        except ValueError:
            print("Invalid input. Please enter a valid number for the first number.")

    # Get the second number from the user
    while True:
        try:
            num2_str = input("Enter the second number: ")
            num2 = float(num2_str) # Convert input to a float
            break # Exit loop if conversion is successful
        except ValueError:
            print("Invalid input. Please enter a valid number for the second number.")

    print("\n--- Results ---")

    # Addition
    sum_result = num1 + num2
    print(f"Addition ({num1} + {num2}) = {sum_result}")

    # Subtraction
    difference_result = num1 - num2
    print(f"Subtraction ({num1} - {num2}) = {difference_result}")

    # Multiplication
    product_result = num1 * num2
    print(f"Multiplication ({num1} * {num2}) = {product_result}")

    # Division (with error handling for division by zero)
    if num2 != 0:
        division_result = num1 / num2
        print(f"Division ({num1} / {num2}) = {division_result}")
    else:
        print("Division by zero is not possible.")

# Call the function to run the program
if __name__ == "__main__":
    perform_calculations()
```

```python
8. Write a program to demonstrate the use of logical operators.¬†
    
Ans. demonstrate_logical_operators():
    """
    Demonstrates the use of 'and', 'or', and 'not' logical operators
    with various conditions.
    """
    print("--- Demonstrating Logical Operators ---")

    # --- Variables for Conditions ---
    is_sunny = True
    is_warm = True
    is_weekend = False
    has_discount = False
    item_in_stock = 10
    user_age = 25

    print(f"\nInitial Conditions:")
    print(f"  Is it sunny? {is_sunny}")
    print(f"  Is it warm? {is_warm}")
    print(f"  Is it the weekend? {is_weekend}")
    print(f"  Is there a discount? {has_discount}")
    print(f"  Items in stock: {item_in_stock}")
    print(f"  User age: {user_age}")

    # --- 1. 'and' Operator ---
    print("\n--- 1. 'and' Operator (Both conditions must be True) ---")

    # Scenario 1: Go to the beach if it's sunny AND warm
    can_go_beach = is_sunny and is_warm
    print(f"Can go to beach (sunny AND warm): {can_go_beach}") # Expected: True

    # Scenario 2: Buy item if it's in stock AND has a discount
    can_buy_item = (item_in_stock > 0) and has_discount
    print(f"Can buy item (in stock AND discount): {can_buy_item}") # Expected: False

    # Scenario 3: Complex 'and' with age and weekend
    eligible_for_special = (user_age >= 18) and is_weekend
    print(f"Eligible for special (age >= 18 AND weekend): {eligible_for_special}") # Expected: False


    # --- 2. 'or' Operator (At least one condition must be True) ---
    print("\n--- 2. 'or' Operator (At least one condition must be True) ---")

    # Scenario 1: Stay home if it's NOT sunny OR NOT warm
    stay_home_bad_weather = (not is_sunny) or (not is_warm)
    print(f"Stay home (NOT sunny OR NOT warm): {stay_home_bad_weather}") # Expected: False (since both are True, not sunny is False, not warm is False, so False or False is False)

    # Scenario 2: Qualify for free shipping if total is high OR it's a special day
    total_high = 120
    is_special_day = True
    free_shipping = (total_high >= 100) or is_special_day
    print(f"Free shipping (total >= 100 OR special day): {free_shipping}") # Expected: True

    # Scenario 3: Allow entry if user is VIP OR has a ticket
    is_vip = False
    has_ticket = True
    allow_entry = is_vip or has_ticket
    print(f"Allow entry (VIP OR has ticket): {allow_entry}") # Expected: True


    # --- 3. 'not' Operator (Inverts the boolean value) ---
    print("\n--- 3. 'not' Operator (Inverts the boolean value) ---")

    # Scenario 1: Check if it's NOT raining
    is_raining = False
    print(f"Is it raining? {is_raining}")
    not_raining = not is_raining
    print(f"Is it NOT raining? {not_raining}") # Expected: True

    # Scenario 2: Check if there is NO stock
    print(f"Items in stock: {item_in_stock}")
    no_stock = not (item_in_stock > 0)
    print(f"Is there NO stock? {no_stock}") # Expected: False

    # Scenario 3: Check if a condition is NOT met
    is_adult = user_age >= 18
    print(f"Is user an adult? {is_adult}")
    not_adult = not is_adult
    print(f"Is user NOT an adult? {not_adult}") # Expected: False


# Call the function to run the demonstration
if __name__ == "__main__":
    demonstrate_logical_operators()
```

```python
9. Write a Python program to convert user input from string to integer, float, and boolean types.

Ans. def convert_user_input():
    """
    Prompts the user for input and attempts to convert it to integer,
    float, and boolean types, showing the results.
    """
    print("--- User Input Type Conversion Demo ---")

    user_input = input("Enter a value: ")
    print(f"\nOriginal Input: '{user_input}' (Type: {type(user_input)})")

    # --- Convert to Integer ---
    print("\n--- Converting to Integer (int()) ---")
    try:
        converted_int = int(user_input)
        print(f"  Successfully converted to integer: {converted_int} (Type: {type(converted_int)})")
    except ValueError:
        print("  Could not convert to integer. Input must be a whole number string (e.g., '123', '-5').")
    except TypeError: # Less common for direct input, but good practice
        print("  TypeError: Cannot convert this type to integer.")

    # --- Convert to Float ---
    print("\n--- Converting to Float (float()) ---")
    try:
        converted_float = float(user_input)
        print(f"  Successfully converted to float: {converted_float} (Type: {type(converted_float)})")
    except ValueError:
        print("  Could not convert to float. Input must be a number string (e.g., '3.14', '-10', '7.0').")
    except TypeError:
        print("  TypeError: Cannot convert this type to float.")

    # --- Convert to Boolean ---
    print("\n--- Converting to Boolean (bool()) ---")
    # bool() conversion is generally always successful but follows specific rules
    converted_bool = bool(user_input)
    print(f"  Converted to boolean: {converted_bool} (Type: {type(converted_bool)})")
    print("  (Remember: Empty strings, 0, None, empty lists/tuples/dicts convert to False; anything else to True.)")


# Run the program
if __name__ == "__main__":
    convert_user_input()
```

```python
10. 10. Write code to demonstrate type casting with list elements.

Ans. # Original list with mixed data types
my_list = ["1", "2", "3", 4, 5.0, "True", "3.14", "false"]
print(f"Original List: {my_list}")
print(f"Types in Original List: {[type(item).__name__ for item in my_list]}\n")

# --- 1. Converting list elements to Integers (int()) ---
# Note: This will fail if elements are not valid whole number strings or actual numbers
int_list = []
for item in my_list:
    try:
        int_list.append(int(item))
    except (ValueError, TypeError):
        int_list.append(f"NotConvertibleToInt ({item})") # Handle non-convertible items

print(f"List after converting to Integers: {int_list}")
print(f"Types in Integer List: {[type(item).__name__ for item in int_list]}\n")

# --- 2. Converting list elements to Floats (float()) ---
# Note: This will fail if elements are not valid number strings or actual numbers
float_list = []
for item in my_list:
    try:
        float_list.append(float(item))
    except (ValueError, TypeError):
        float_list.append(f"NotConvertibleToFloat ({item})") # Handle non-convertible items

print(f"List after converting to Floats: {float_list}")
print(f"Types in Float List: {[type(item).__name__ for item in float_list]}\n")

# --- 3. Converting list elements to Strings (str()) ---
# This is generally always successful as most data can be represented as a string
str_list = [str(item) for item in my_list] # Using a list comprehension for conciseness

print(f"List after converting to Strings: {str_list}")
print(f"Types in String List: {[type(item).__name__ for item in str_list]}\n")

# --- 4. Converting list elements to Booleans (bool()) ---
# Remember bool() conversion rules: empty strings, 0, None, empty collections are False
bool_list = [bool(item) for item in my_list] # Using a list comprehension

print(f"List after converting to Booleans: {bool_list}")
print(f"Types in Boolean List: {[type(item).__name__ for item in bool_list]}\n")

# --- 5. Example of converting a list of strings to a list of numbers for calculation ---
price_strings = ["10.50", "25", "5.75"]
print(f"Original Price Strings: {price_strings}")

# Convert to floats to perform summation
prices_as_floats = [float(price) for price in price_strings]
total_price = sum(prices_as_floats)

print(f"Prices as Floats: {prices_as_floats}")
print(f"Total Price (sum of floats): {total_price}\n")
```

```python
11. Write a program that checks if a number is positive, negative, or zero.

Ans. def check_number_type():
    """
    Prompts the user for a number and determines if it's positive, negative, or zero.
    """
    print("--- Number Type Checker ---")

    while True:
        try:
            user_input = input("Enter a number: ")
            number = float(user_input) # Use float to handle decimals and zero correctly
            break # Exit loop if a valid number is entered
        except ValueError:
            print("Invalid input. Please enter a valid number.")

    if number > 0:
        print(f"The number {number} is positive. üëç")
    elif number < 0:
        print(f"The number {number} is negative. üëé")
    else: # This 'else' catches the case where number is not > 0 and not < 0, meaning it must be 0
        print(f"The number {number} is zero. 0Ô∏è‚É£")

# Run the program
if __name__ == "__main__":
    check_number_type()
```

```python
14. Write a program to reverse a string using a while loop.¬†

Ans. def reverse_string_while_loop(s):
    """
    Reverses a given string using a while loop.

    Args:
      s: The input string to be reversed.

    Returns:
      The reversed string.
    """
    reversed_s = ""
    index = len(s) - 1  # Start from the last character's index

    while index >= 0:
        reversed_s += s[index]  # Append the character at the current index
        index -= 1              # Move to the previous character

    return reversed_s

# --- Main part of the program ---
if __name__ == "__main__":
    user_input = input("Enter a string to reverse: ")
    reversed_result = reverse_string_while_loop(user_input)
    print(f"Original string: '{user_input}'")
    print(f"Reversed string: '{reversed_result}'")

    print("\n--- Testing with examples ---")
    print(f"Reversing 'Python': '{reverse_string_while_loop('Python')}'")
    print(f"Reversing 'hello': '{reverse_string_while_loop('hello')}'")
    print(f"Reversing '12345': '{reverse_string_while_loop('12345')}'")
    print(f"Reversing '': '{reverse_string_while_loop('')}'") # Empty string
    print(f"Reversing 'a': '{reverse_string_while_loop('a')}'")  # Single character
```

```python
15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

Ans. def calculate_factorial_while_loop():
    """
    Calculates the factorial of a non-negative integer provided by the user
    using a while loop.
    """
    print("--- Factorial Calculator (using while loop) ---")

    while True:
        try:
            num_str = input("Enter a non-negative integer: ")
            num = int(num_str)

            if num < 0:
                print("Factorial is not defined for negative numbers. Please enter a non-negative integer.")
            else:
                break # Exit loop if input is a valid non-negative integer
        except ValueError:
            print("Invalid input. Please enter a whole number.")

    # Calculate factorial using a while loop
    factorial = 1
    # If the number is 0, factorial is 1, and the loop won't run as num is not > 0
    current_num = num

    while current_num > 0:
        factorial *= current_num # Multiply factorial by current_num
        current_num -= 1         # Decrement current_num

    print(f"The factorial of {num} is {factorial}")

# Run the program
if __name__ == "__main__":
    calculate_factorial_while_loop()
```
