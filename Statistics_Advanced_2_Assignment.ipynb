### Question 1: What is hypothesis testing in statistics?

Hypothesis testing is a statistical method used to make decisions about a population based on sample data. It involves formulating a null hypothesis (H0) and an alternative hypothesis (H1), then using sample data to determine whether to reject the null hypothesis. The process includes selecting a significance level, calculating a test statistic, and comparing it to a critical value or using a p-value to make a decision.

### Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

The null hypothesis (H0) is a statement that there is no effect or no difference, and it serves as the default assumption. The alternative hypothesis (H1) is what you want to prove, indicating the presence of an effect or difference. Hypothesis testing evaluates whether the sample data provides enough evidence to reject H0 in favor of H1.

### Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

The significance level (denoted as alpha, α) is the probability of rejecting the null hypothesis when it is actually true. Common values are 0.05 or 0.01. It defines the threshold for the p-value; if the p-value is less than α, the null hypothesis is rejected. It helps control the Type I error rate in hypothesis testing.

### Question 4: What are Type I and Type II errors? Give examples of each.

Type I error occurs when the null hypothesis is rejected when it is actually true (false positive). Example: Concluding a new drug is effective when it is not.
Type II error occurs when the null hypothesis is not rejected when it is false (false negative). Example: Concluding a new drug is not effective when it actually is.

### Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

A Z-test is used when the population standard deviation is known and the sample size is large (n > 30). A T-test is used when the population standard deviation is unknown and the sample size is small (n ≤ 30). Both tests compare sample data to a population parameter to determine statistical significance.

### Question 6: Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

n = 10
p = 0.5
size = 1000

data = np.random.binomial(n, p, size)

plt.hist(data, bins=range(n+2), edgecolor='black', align='left')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of successes')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python.

```python
import numpy as np
from scipy.stats import norm

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mu = 50
sigma = 1
n = len(sample_data)
mean_sample = np.mean(sample_data)

z = (mean_sample - mu) / (sigma / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))

print(f"Z-statistic: {z:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

### Question 8: Simulate data from a normal distribution and calculate the 95% confidence interval for its mean.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

data = np.random.normal(loc=100, scale=15, size=100)
mean = np.mean(data)
sem = stats.sem(data)
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=sem)

print(f"Mean: {mean:.2f}")
print(f"95% Confidence Interval: {ci}")

plt.hist(data, bins=15, edgecolor='black')
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Question 9: Calculate Z-scores from a dataset and visualize the standardized data using a histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(100, 15, 100)
z_scores = (data - np.mean(data)) / np.std(data)

plt.hist(z_scores, bins=15, edgecolor='black')
plt.title('Z-scores Histogram')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

print("Z-scores represent how many standard deviations a data point is from the mean.")
```
